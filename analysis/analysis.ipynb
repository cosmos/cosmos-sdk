```python
from analysis.read_otel import load_otel_runs
from analysis.analysis import block_summary, plot_block_durations
import pandas as pd
```

```python
runs = load_otel_runs("/Users/arc/iavl-bench-data/sims2")
```

```python
summary_data = []
for name, run in runs.items():
    summary = block_summary(run)
    summary_data.append({
        'run': name,
        'duration': summary.total_duration_seconds,
        'blocks': summary.block_count,
        'avg_block_time': summary.total_duration_seconds / summary.block_count,
    })
pd.DataFrame(summary_data).set_index('run')
```
